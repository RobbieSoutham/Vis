```python
import pandas as pd
data = pd.read_csv("../data/famd_with_cluster_means.csv")[["Age", "Component 0", "Component 1", "Body mass index", "ID", "Cluster", "Work load Average/day ", "Social drinker", "Social smoker", "Absenteeism time in hours"]]
```

```python
pd.set_option('display.max_rows', 5000)
```

```python
data = data.drop(columns=['id'])
```

```python
counts = data.groupby('ID').size().to_frame().rename(columns={0: "counts"})
```

```python
total_times = data.groupby('ID').sum()['Absenteeism time in hours'].to_frame().rename(columns={'Absenteeism time in hours': "total absence time"})
```

```python
means = data.groupby('Cluster').mean().drop(columns=["ID"]).add_suffix("_m")
```

```python
new = data.join(counts, on="ID").join(total_times, on="ID")
new = new.join(means, on="Cluster")
```

new["Social smoker"] = new["Social smoker"].replace(0, "Non smoker").replace(1, "Social smoker")
new["Social drinker"] = new["Social drinker"].replace(0, "Non drinker").replace(1, "Social drinker")

```python
new.to_csv("../data/final.csv")
```
